```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
iris=pd.read_csv('IRIS.csv')
iris.head(5)
```

```python
numeric_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
correlation_matrix = iris[numeric_cols].corr()

plt.figure(figsize=(10, 8))

sns.heatmap(correlation_matrix,
            annot=True,
            fmt='.2f',
            cmap='RdYlBu_r',
            center=0,
            square=True,
            linewidths=0.5,
            cbar_kws={'label': 'Correlation Coefficient'})

plt.title('Correlation Matrix')

plt.tight_layout()
plt.show()
```

```python
# Create mask for upper triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

plt.figure(figsize=(10, 8))

sns.heatmap(correlation_matrix,
            mask=mask,
            annot=True,
            fmt='.2f',
            cmap='coolwarm',
            center=0,
            square=True,
            linewidths=0.5,
            annot_kws={'size': 12})

plt.title('Correlation Matrix (Lower Triangle)', fontsize=14, fontweight='bold', pad=20)

plt.tight_layout()
plt.show()
```

```python
g = sns.clustermap(iris[numeric_cols].corr(),
                   annot=True,
                   fmt='.2f',
                   cmap='viridis',
                   linewidths=0.5,
                   figsize=(12, 10),
                   standard_scale=1,  # Standardize columns
                   cbar_pos=(0.02, 0.8, 0.03, 0.15))

g.fig.suptitle('Clustered Heatmap', fontsize=14, fontweight='bold', y=1.02)

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=iris, x="species", y="petal_length")

plt.title("Petal Length vs Species")
plt.xlabel("Species")
plt.ylabel("Petal Length")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue="species", style="species", s=80)

plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
species_counts = iris['species'].value_counts()
plt.pie(
    species_counts,
    labels=species_counts.index,
)
plt.title("Distribution of Iris Species")
plt.show()
```

```python
g = sns.pairplot(iris, hue='species', palette='husl',
                 diag_kind='kde', markers=['o', 's', 'D'],
                 plot_kws={'alpha': 0.6})

g.fig.suptitle('Iris Dataset: Pairwise Relationships', fontsize=14, fontweight='bold', y=1.02)

plt.show()
```
